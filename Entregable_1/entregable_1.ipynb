Ahora ya que solo queremos los datos del ultimo cuatri, eliminaremos todos los datos restantes.

```python
df_cuar = df['q'] == '2022q1'
df_el = df[df_cuar]
```

Siguendo con nuestra limpieza de datos ahora eliminaremos algunas columnas que no nos seran de ayuda a la hora de querer sacar
los calculos.

```python
df_el = df_el.drop(['Ticker', 'Name', 'partind', 'Sector NAICS\nlast available', 'Sector\nEconomatica', 'Class', 'Exchange / Src', 'extraincome', 'shortdebt', 'stockholderequity', 'longdebt'], axis = 1)
```

```python
df_el['R'] = np.log(df.groupby(['firm'])['adjprice'].shift(-1)) - np.log(df.groupby(['firm'])['adjprice'].shift(3))
```

Realizamos los calculos para OPM, EPSP y Book to Market.

```python
df_el['Book'] = df_el['totalassets'] - df_el['totalliabilities']
df_el['Market'] = df_el['originalprice'] * df_el['sharesoutstanding']
df_el['Book_Market'] = df_el['Book'] / df_el['Market']
df_el['EBIT'] = df_el['revenue'] - df_el['cogs'] - df_el['sgae'] - df_el['otheropexp']
df_el['NetIncome'] = df_el['EBIT'] - df_el['incometax'] - df_el['finexp']
df_el['EPS'] = df_el['NetIncome'] / df_el['sharesoutstanding']
df_el['OPM'] = df_el['EBIT'] / df_el['revenue']
df_el['EPSP'] = df_el['EPS'] / df_el['originalprice']
```

Ahora asignaremos los dummys para poder obtner las empresas.

```python
df_cl = df_el.sort_values(by = 'Market')
```

```python

df_cl["Small"] = df_cl.index <= len(df_cl) / 3
df_cl["Small"].astype(int)
```

```python
df_cl["Medium"] = (df_cl.index <= (2 * len(df_cl) / 3)) & (df_cl.index > (len(df_cl) / 3))
df_cl["Medium"].astype(int)
```

Para poder iniciar con nuestra matriz de varianza y covariana vamos a separar nuestras variables en x y y.

```python
x = df_cl[['OPM', 'EPSP', 'Book_Market', 'Small', 'Medium']]
y = df_cl['R']
```

Seguimos paso por paso la formula para sacar la matriz de varianza y covarianza.

```python
var_cor = x.to_numpy()
z = np.transpose(var_cor)
u = np.zeros((z.shape[1], 1)) + 1
m = np.dot(z, var_cor)
n = np.dot(z, u)
t = np.transpose(n)
p = np.dot(n, t)
q = p / var_cor.shape[0]
l = m - q
matriz = l / (var_cor.shape[0] - 1)
pd.DataFrame(matriz)
x.cov()
```

Para ahora obtener la matriz de correlaciones, de igual manera siguiendo la formula paso a paso.

```python
b = np.diagonal(matriz)
stan = np.sqrt(b)
n = [[ 0 for y in range( len(stan) ) ] for x in range( len(stan) )]
for i in range(len(stan)):
  for j in range(len(stan)):
    n[i][j] = stan[i] * stan[j]
correlation = [[ 0 for y in range( len(n) ) ] for x in range( len(n) )]
for i in range(len(n)):
  for j in range(len(n)):
    correlation[i][j] = matriz[i][j] / n[i][j]
x.corr()
```
